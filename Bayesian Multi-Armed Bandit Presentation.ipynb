This notebook solves the Multi-Armed Bandit problem using Bayesian updates on 
Thompson sampling of beta distributions. This method focuses on finding which 
option is best quickly and exploiting it, balancing exploration and exploitation.


```python
""" This cell provides functions that are useful later
but not worth talking about. """

# A non-blocking print function.
# Prepending '\r' to a string allows same-line replacement
import sys
def nbprint(x):
    sys.stdout.write(x)
    sys.stdout.flush()
    

# Round a number to n significant digits
from math import log10, floor
def round_to_n(x, n):
    if x == 0:
        return 0
    return round(x, -int(floor(log10(abs(x)))) + (n - 1))


# Return the mode of a beta distribution
# Note that this is just wins/trials, since the ignorance prior
# for 0 trials is a=1, b=1, giving wins=a-1, losses=b-1, trials=a+b-2
def beta_mode(a, b):
    # mode is (a-1)/(a+b-2)
    if (a+b-2) == 0:
        return 0.5
    else:
        return (a-1)/(a+b-2)

    
# Render a list as an html table. 
# Use make_list_table for an easy input interface to this class
class ListTable(list):
    """ Overridden list class which takes a 2-dimensional list of 
        the form [[1,2,3],[4,5,6]], and renders an HTML Table in 
        IPython Notebook. """
    
    def _repr_html_(self):
        html = ["<table>"]
        for row in self:
            html.append("<tr>")
            
            for col in row:
                html.append("<td>{0}</td>".format(col))
            
            html.append("</tr>")
        html.append("</table>")
        return ''.join(html)    

    
# Uses the ListTable class to create an HTML table from lists.
# 'titles' should be a list of strings that label the columns,
# and the 'datalists' arguments should be lists of data, one per column
# Currently configured for displaying floats
# EXAMPLE:
# x = [1,2,3]; y=[4,5,6]
# table = make_list_table(['x','y'], x, y)
# table
def make_list_table(titles, *datalists):
    table = ListTable()
    table.append(titles)
    for i in range(len(datalists[0])):
        table.append(['{:.3f}'.format(x[i]) for x in datalists])
    return table
```

```python
"""
Monte-Carlo estimation of the chance of each candidate being the best
"""

# Generate samples from each distribution
num_candidates = len(a)
num_samples = 100000
samples = np.zeros(shape=(num_samples,num_candidates), dtype=float)
for i in range(num_candidates):
    samples[:,i]  = ss.beta(a[i],b[i]).rvs(size=num_samples)
    
# See what percentage of the samples from each win
winner = np.zeros(shape=(num_samples,num_candidates,), dtype=bool)
for i, row in enumerate(samples):
    winner[i,np.argmax(row)] = True
prob_winner = np.sum(winner,0)/num_samples
leader_id = np.argmax(prob_winner)

print('Odds of being the best candidate: ')
for i in range(num_candidates):
    print('C{:<2d}: {:>5.1f}%'.format(i+1, prob_winner[i]*100))    
```

```python
"""
These classes handle beta distribution updates for the bandits.
"""

class bandit_estimator_betadist(object):
    def __init__(self, num_candidates=2, prior=(1.0,1.0)):
        self.num_candidates = num_candidates
        self.prior = prior
        self.trials = np.zeros(shape=(num_candidates,), dtype=int)
        self.wins = np.zeros(shape=(num_candidates,), dtype=int)
        self.leader_id = None
        
    def update_on_trial(self, candidate_id, success):
        self.trials[candidate_id] += 1
        if (success):
            self.wins[candidate_id] += 1
            
    def pick_candidate_for_trial(self):
        samples = []
        for i in range(self.num_candidates):
            dist = ss.beta(self.prior[0]+self.wins[i],
                        self.prior[1]+self.trials[i]-self.wins[i])
            samples += [ dist.rvs() ]
        return samples.index( max(samples) )
    
    def conf_interval(self, conf, trials, wins):
        return ss.beta.interval(conf, self.prior[0]+wins, self.prior[1]+trials-wins)
    
    def candidate_mode(self, candidate_id):
        # mode is (a-1)/(a+b-2)
        # a = prior[0]+wins, b = prior[1]+trials-wins
        if (self.prior[0]+self.prior[1]+self.trials[candidate_id]-2) == 0:
            return 0.5
        else:
            return ( (self.prior[0]+self.wins[candidate_id]-1) / 
                     (self.prior[0]+self.prior[1]+self.trials[candidate_id]-2) )
        
    def generate_samples(self, num_samples, num_trials, num_wins):
        dist = ss.beta(self.prior[0]+num_wins,
                       self.prior[1]+num_trials-num_wins)
        return dist.rvs(size=num_samples) 
    
    def calc_value_left(self, samples, conf):
        # given sampled distributions, calculate 95th percentile of remaining 
        #   value distribution for further trials
        num_samples = len(samples)
        regrets = np.zeros(shape=(num_samples,), dtype=float)
        for i, sample in enumerate(samples):
            regrets[i] = max(sample) - sample[self.leader_id]
        bins = 100000  # even 1/100000 spacing
        hist, bin_edges = np.histogram(regrets, bins=bins)
        idx = 0
        cum_value = np.cumsum(hist)/num_samples
        for i in range(bins):
            if cum_value[i] >= conf:
                idx = i+1    # bin edges are left-indexed
                break
        return bin_edges[idx], regrets
        
    def evaluate_progress(self, num_samples, conf):
        # generate samples of the candidate estimate distributions
        samples = np.zeros(shape=(num_samples,self.num_candidates,), dtype=float)
        for i in range(num_candidates):
            samples[:,i]  = self.generate_samples(num_samples, self.trials[i], self.wins[i])
    
        # calculate probability that each candidate is the best as a percentage
        winner = np.zeros(shape=(num_samples,self.num_candidates,), dtype=bool)
        for i, row in enumerate(samples):
            winner[i,np.argmax(row)] = True
        prob_winner = np.sum(winner,0)/num_samples
        self.leader_id = np.argmax(prob_winner)
        
        # calculate (conf) percentile of remaining value distribution for further trials
        value_left, regrets = self.calc_value_left(samples, conf)
                
        return prob_winner, value_left
                       
class bandit(object):
    def __init__(self, p=0.01):
        self.p = p
       
    def trial(self):
        return np.random.random() < self.p
```

```python
"""
Set up experiment candidates, do trial runs and check confidence and 
expected relative improvement every (evaluation_interval) trials until
one of the thresholds is met
"""

from IPython.display import display, clear_output

# Define parameters of the test
theta = np.array([0.20, 0.22])
min_trials = 2000
max_trials = 10000
evaluation_interval = 100
conf = 0.95
val_threshold = 0.01

candidates = [bandit(x) for x in theta]
num_candidates = len(candidates)
bandit_est = bandit_estimator_betadist(num_candidates, (1.0,1.0))

cum_trials_hist = np.zeros(shape=(max_trials,num_candidates), dtype=int)
cum_wins_hist = np.zeros(shape=(max_trials,num_candidates), dtype=int)
leader_id = np.zeros(1)
prob_winner = np.zeros(shape=(num_candidates,))
value_left = np.zeros(1)

# set up graph
x = np.arange(0., 1., 0.0001)
fig, ax = plt.subplots(1)

for i in range(max_trials):
    # Pick the next trial candidate, do the trial, store the result
    candidate_id = bandit_est.pick_candidate_for_trial()
    result = candidates[candidate_id].trial()

    # Update distribution of trial candidate
    bandit_est.update_on_trial(candidate_id, result)

    # Store historical values for graphing 
    cum_trials_hist[i] = bandit_est.trials
    cum_wins_hist[i] = bandit_est.wins
    
    if (i+1) % evaluation_interval == 0:
        # Every thousand loops, evaluate progress
        prob_winner, value_left = bandit_est.evaluate_progress(100000, 0.95)
        leader_id = np.argmax(prob_winner)
        
        # update graph
        a = bandit_est.wins+1
        b = bandit_est.trials-bandit_est.wins+1
        
        clear_output(wait=True)
        ax.cla()

        plt.xlabel('probability')
        plt.ylabel('magnitude of probability density')
        plt.title('Candidate distributions')
        plt.xlim([0.05,0.45])
        plt.legend(loc=2)
        plt.gcf().set_size_inches(12,6)
        ax._get_lines.set_color_cycle(None)  # Reset the color cycle
        
        for j in range(num_candidates):
            nextcolor = next(ax._get_lines.color_cycle)
            plt.plot(x, ss.beta.pdf(x,a[j],b[j]), color=nextcolor, 
                     label='{} wins, {} losses'.format(a[j]-1, b[j]-1))
            plt.axvline(theta[j], color=nextcolor, linewidth=2, 
                     label='C{} true rate'.format(j+1))
            plt.legend(loc=2)
        display(fig)
        
        # Output progress indicator
        nbprint('\r{} trials, probabilities {}, 95th %ile of value left'
                ' is {:.2g}'.format(i+1, prob_winner, value_left))
     
        # Quit when over 95% certain of a winner or over 95% sure that less than 
        #   1% relative improvement remains
        if (((max(prob_winner) > conf) or (value_left <= val_threshold*bandit_est.candidate_mode(leader_id)))
            and (i+1) >= min_trials):
            break

# Clean up some values for later analysis
num_trials = i+1
cum_trials_hist = cum_trials_hist[:num_trials]
cum_wins_hist = cum_wins_hist[:num_trials]
leader_winrate = bandit_est.candidate_mode(leader_id)
rel_value_left = value_left / leader_winrate

print('\nC{} wins with {:.3g}% conf., there is 95% conf.'
      ' that <{:.2g}% relative improvement remains.'.format(leader_id+1,
                                                            prob_winner[leader_id]*100,
                                                            rel_value_left*100,))

# prevent duplicate plot from appearing
plt.close()
```

```python
""" Run the algorithm many times and see how often the answer is correct """

num_runs = 1000
trials_taken = np.zeros(shape=(num_runs,), dtype=int)
run_correct = np.zeros(shape=(num_runs,), dtype=bool)
run_endstate = np.zeros(shape=(num_runs,), dtype=int)
trials_until_corrected = np.zeros(shape=(num_runs,), dtype=int)
# 0 = has not ended
# 1 = ended due to confidence 
# 2 = ended due to value remaining
# 3 = ended due to number of trials taken

for run in range(num_runs):
    theta = np.array([0.50, 0.55])
    candidates = [bandit(x) for x in theta]
    num_candidates = len(candidates)

    bandit_est = bandit_estimator_betadist(num_candidates, (1.0,1.0))

    min_trials = 0
    max_trials = 100000
    evaluation_interval = 100
    conf = 0.95
    val_threshold = 0.01
    leader_id = np.zeros(1)
    prob_winner = np.zeros(shape=(num_candidates,))
    value_left = np.zeros(1)

    for i in range(max_trials):
        # Pick the next trial candidate, do the trial, store the result
        candidate_id = bandit_est.pick_candidate_for_trial()
        result = candidates[candidate_id].trial()

        # Update distribution of trial candidate
        bandit_est.update_on_trial(candidate_id, result)
   
        if (i+1) % evaluation_interval == 0 and (i+1) >= min_trials:
            # Every thousand loops, evaluate progress
            prob_winner, value_left = bandit_est.evaluate_progress(100000, 0.95)
            leader_id = np.argmax(prob_winner)
            
            # Output progress indicator
            nbprint('\rrun {}, {} trials, probabilities {}, 95th %ile of value left'
                    ' is {}'.format(run+1, i+1, prob_winner, round_to_n(value_left,2)))
     
            # Quit when over 95% certain of a winner or over 95% sure that less than 
            #   1% relative improvement remains.
            # Modified to allow incorrect runs to continue until correct,
            #   logging how long it takes for them to become correct.
            if ((max(prob_winner) > conf) 
              or (value_left <= val_threshold*bandit_est.candidate_mode(leader_id))):
                if run_endstate[run] == 0:
                    if max(prob_winner) > 0.95:
                        run_endstate[run] = 1
                    elif value_left <= 0.01*bandit_est.candidate_mode(leader_id):
                        run_endstate[run] = 2
                    trials_taken[run] = i+1
                    if leader_id == np.argmax(theta):
                        run_correct[run] = True
                if leader_id != np.argmax(theta):
                    trials_until_corrected[run] = i+1+evaluation_interval
                else:
                    break
                
    if run_endstate[run] == 0:
        run_endstate[run] = 3
        trials_until_corrected[run] = max_trials

print('\n{} correct, {} incorrect'.format(np.sum(run_correct), num_runs-np.sum(run_correct)))
```
